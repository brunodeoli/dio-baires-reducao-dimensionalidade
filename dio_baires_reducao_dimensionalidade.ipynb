```python
import os
import matplotlib.pyplot as plt

from PIL import Image
```

```python
img = Image.open('luna.jpeg')
print(f"Original size: {img.width} x {img.height}")
```

```python
def resize(image, width, height):
  image.resize((width, height))
  return image

def to_grayscale(image):
  pixels_rgb = list(image.getdata())
  image = Image.new("L", (image.width, image.height))
  pixels_gray = [(int(0.299*r + 0.587*g + 0.114*b),) for r, g, b in pixels_rgb]
  image.putdata([p[0] for p in pixels_gray])
  return image

def binarize(image, threshold=127):
  pixels_gray = list(image.getdata())
  pixels_bin = [(255,) if gray > threshold else (0,) for gray in pixels_gray]
  image.putdata([p[0] for p in pixels_bin])
  return image
```

```python
#RESIZE IMAGE
img = img.resize((240, 320))
img
```

```python
#APPLY GRAYSCALE FILTER TO THE IMAGE
img_grayscale = to_grayscale(img)
img_grayscale
```

```python
#APPLY BINARIZE FILTER TO THE IMAGE
img_binarize = binarize(img_grayscale, 100)
img_binarize
```
